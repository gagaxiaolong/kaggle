```python
import json

from tqdm import tqdm

DATA_PATH = 'data/'
```

```python
with open(DATA_PATH+'internal-citations.json') as f:
    citations = json.loads(f.read())
def get_paper_citation_count(paper_id):
    try:
        citation_count = len(citations[paper_id])
    except:
        citation_count = 0
    return citation_count
        
def get_papers():
    papers = {}
    with open(DATA_PATH+'arxiv-metadata-oai-snapshot.json') as f:
        for line in tqdm(f):
            paper=json.loads(line)
            papers[paper['id']] = {
                'title': paper['title'],
                'citation_count': get_paper_citation_count(paper['id']) 
#                 'categories': paper['categories']
            }
    return papers

papers = get_papers()
```

```python
TOP10 = []
top10_citations = []
count = 0
for key, paper in papers.items():
    if len(TOP10) < 10:
        TOP10.append(paper)
        top10_citations.append(paper['citation_count'])
    else:
        if paper['citation_count'] > min(top10_citations):
            top10_citations.remove(min(top10_citations))
            for paper in TOP10:
                if paper['citation_count'] == min(top10_citations):
                    TOP10.remove(paper)
            top10_citations.append(paper['citation_count'])
            TOP10.append(paper)
    count += 1
    if count % 1000 == 0:
        print(TOP10)
```
